```python
import numpy as np

# Changed input vector
x = np.array([0.5, 0.9, 0.1])

# Changed weight matrix (3 inputs -> 2 neurons)
w = np.array([
    [0.4, 0.1],
    [0.7, 0.3],
    [0.2, 0.8]
])

# Changed bias
b = np.array([0.05, 0.02])

# Linear combination
z = np.dot(x, w) + b
print("Output (z): ", z)

# Sigmoid activation
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

a = sigmoid(z)
print("Activated Output: ", a)
```

```python
w2= np.array([[0.8],[0.4]])
b2= np.array([0.02])
z2= np.dot(a,w2)+b
a2=sigmoid(z2)
print("Final Output: ", a2)
```

```python
def relu(x):
  return np.maximum(0, x)

a_relu = relu(z)
print("ReLU Activated Output: ", a_relu)
a2_relu = relu(z2)
print("ReLU Final Output: ", a2_relu)
```
